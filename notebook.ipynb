## Visualization & Explanatory Analysis

Pertanyaan Bisnis 1:

# Bagaimana pengaruh kondisi cuaca terhadap jumlah penyewaan sepeda?

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.barplot(data=day_df, x="weathersit", y="cnt")
plt.title("Rata-rata Penyewaan Berdasarkan Cuaca")
plt.show()
```

INSIGHT

Berdasarkan visualisasi, kondisi cuaca memiliki pengaruh terhadap jumlah penyewaan sepeda. Cuaca yang lebih baik cenderung menghasilkan jumlah penyewaan yang lebih tinggi, sedangkan cuaca yang kurang baik menyebabkan jumlah penyewaan menurun.


### Pertanyaan 2:
# Bagaimana pengaruh musim terhadap jumlah penyewaan sepeda?

```python
season_labels = {
    1: "Spring",
    2: "Summer",
    3: "Fall",
    4: "Winter"
}

day_df["season_name"] = day_df["season"].map(season_labels)

plt.figure(figsize=(8,5))
sns.barplot(data=day_df, x="season_name", y="cnt")
plt.title("Jumlah Penyewaan Berdasarkan Musim")
plt.show()
```

**Insight:**
- Berdasarkan visualisasi, terdapat perbedaan jumlah penyewaan sepeda pada setiap musim. Beberapa musim menunjukkan tingkat penyewaan yang lebih tinggi dibandingkan musim lainnya, yang menunjukkan bahwa faktor musiman berpengaruh terhadap penggunaan sepeda.
